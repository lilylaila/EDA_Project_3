Acquire the Data

```r
BostonURL <- "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
Boston <- read.csv(url(BostonURL), sep = "", header = FALSE)
Boston
```

```r
colnames(Boston) <- c("CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO", "B", "LSTAT", "MEDV")
Boston
```

```r
dim(Boston)
```

```r
str(Boston)
```

```r
summary(Boston)
```

```r
Boston$CHAS = factor(Boston$CHAS)
Boston$RAD = factor(Boston$RAD)
Boston_f = Filter(is.numeric, Boston)
```

```r
sum_vals = data.frame(feature=colnames(Boston_f))
sum_vals['mean_'] = sapply(Boston_f, mean)
sum_vals
```

```r
library(reshape2)
library(ggplot2)

ggplot(melt(sum_vals), aes(x = feature, y = value, fill = variable)) + 
    geom_bar(stat = "identity", width=0.5, position = "dodge")
```

```r
library(dplyr, warn.conflicts = FALSE)
samples = list()
for (i in 1:10) {
    samples[[i]] = sample_n(Boston_f, 5)
}
```

```r
for (i in 1:10) {
    sum_vals[paste('mean_', i)] = sapply(samples[[i]], mean)
}
```

```r
sum_vals
```

```r
library(reshape2)
library(ggplot2)

ggplot(melt(sum_vals), aes(x = feature, y = value, fill = variable)) + 
    geom_bar(stat = "identity", width=0.5, position = "dodge")
```
