```r
BostonURL <- "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
Boston <- read.csv(url(BostonURL), sep = "", header = FALSE)
colnames(Boston) <- c("CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO", "B", "LSTAT", "MEDV")
Boston
```

```r
dim(Boston)
```

```r
Boston$CHAS = factor(Boston$CHAS)
Boston$RAD = factor(Boston$RAD)
Boston_f = Filter(is.numeric, Boston)
```

```r
install.packages('rpart')
```

```r
library(caret)
library(rpart)
```

```r
calculate_r_2 <- function(actual, prediction) {
    return (1 - (sum((actual-prediction)^2)/sum((actual-mean(actual))^2)))
}

calculate_r_2_for_feature <- function(data, feature) {
    n <- nrow(data)
    
    train_index <- sample(seq_len(n), size = 0.8*n)

    train <- data[train_index,]
    test <- data[-train_index,]
    
    this_formula = paste(feature,"~.")
    fit <- rpart(data=train, formula=as.formula(this_formula))

    y_test <- as.vector(test[[feature]])
    test[feature] <- NULL
    predictions <- predict(fit, test)
    return (calculate_r_2(y_test, predictions))
}
```

```r
calculate_r_2_for_feature(Boston_f,'CRIM')
```

```r
print(paste("CRIM: ", calculate_r_2_for_feature(Boston_f,'CRIM')))
print(paste("ZN: ", calculate_r_2_for_feature(Boston_f,'ZN')))
print(paste("INDUS: ", calculate_r_2_for_feature(Boston_f,'INDUS')))
#print(paste("CHAS: ", calculate_r_2_for_feature(Boston_f,'CHAS')))
print(paste("NOX: ", calculate_r_2_for_feature(Boston_f,'NOX')))
print(paste("RM: ", calculate_r_2_for_feature(Boston_f,'RM')))
print(paste("AGE: ", calculate_r_2_for_feature(Boston_f,'AGE')))
print(paste("DIS: ", calculate_r_2_for_feature(Boston_f,'DIS')))
#print(paste("RAD: ", calculate_r_2_for_feature(Boston_f,'RAD')))
print(paste("TAX: ", calculate_r_2_for_feature(Boston_f,'TAX')))
print(paste("PTRATIO: ", calculate_r_2_for_feature(Boston_f,'PTRATIO')))
print(paste("B: ", calculate_r_2_for_feature(Boston_f,'B')))
print(paste("LSTAT: ", calculate_r_2_for_feature(Boston_f,'LSTAT')))
```

I was unable to run the analysis for CHAS and RAD because I changed those variable from integers to factors.
Based on r_2 test, ZN, NOX, TAX, and DIS have the stronget relationship since these variables are closer to 1.


```r
mean_r2_for_feature <- function (Boston_f, feature) {
    scores = c()
    for (i in 1:100) {
        scores = c(scores, calculate_r_2_for_feature(Boston_f, feature))
    }
    
    return (mean(scores))
}
```

```r
print(paste("CRIM: ", mean_r_2_for_feature(Boston_f,'CRIM')))
print(paste("ZN: ", mean_r_2_for_feature(Boston_f,'ZN')))
print(paste("INDUS: ", mean_r_2_for_feature(Boston_f,'INDUS')))
#print(paste("CHAS: ", mean_r_2_for_feature(Boston_f,'CHAS')))
print(paste("NOX: ", mean_r_2_for_feature(Boston_f,'NOX')))
print(paste("RM: ", mean_r_2_for_feature(Boston_f,'RM')))
print(paste("AGE: ", mean_r_2_for_feature(Boston_f,'AGE')))
print(paste("DIS: ", mean_r_2_for_feature(Boston_f,'DIS')))
#print(paste("RAD: ", mean_r_2_for_feature(Boston_f,'RAD')))
print(paste("TAX: ", mean_r_2_for_feature(Boston_f,'TAX')))
print(paste("PTRATIO: ", mean_r_2_for_feature(Boston_f,'PTRATIO')))
print(paste("B: ", mean_r_2_for_feature(Boston_f,'B')))
print(paste("LSTAT: ", mean_r_2_for_feature(Boston_f,'LSTAT')))
```
