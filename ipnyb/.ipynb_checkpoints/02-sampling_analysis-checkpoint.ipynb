```r
BostonURL <- "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
Boston <- read.csv(url(BostonURL), sep = "", header = FALSE)
Boston
```

```r
colnames(Boston) <- c("CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO", "B", "LSTAT", "MEDV")
head(Boston)
```

```r
dim(Boston)
```

```r
Boston$CHAS = factor(Boston$CHAS)
Boston$RAD = factor(Boston$RAD)
Boston_f = Filter(is.numeric, Boston)
```

Basic Stats

```r
install.packages('moments')
```

```r
library(moments)
```

```r
Boston$CHAS = factor(Boston$CHAS)
Boston$RAD = factor(Boston$RAD)
Boston_f = Filter(is.numeric, Boston)
```

```r
skewness(Boston_f)
```

Most variables have a positive skew except AGE, PTRATIO AND B, which have a negative skew. The latter variables have means that are less than the median.

```r
stats = data.frame(feature=colnames(Boston_f))
stats['mean_'] = sapply(Boston_f, mean)
stats['sd_'] = sapply(Boston_f, sd)
stats['skewness_'] = sapply(Boston_f, skewness)
stats
```

Sampling the Dataset

```r
library(dplyr, warn.conflicts = FALSE)
```

```r
sample = sample_n(Boston_f, 3)
```

```r
sample
```

```r
stats
```

```r
sample_1pct_1 = sample_n(Boston_f, 5)
```

```r
colMeans(sample_1pct_1)
```

```r
colMeans(sample_1pct_1) - stats$mean_
```

Compared to the actual mean, the TAX variable is most below the actual mean, and the B variable has the highest value above the mean.

```r
(colMeans(sample_1pct_1) - stats$mean_)/stats$sd_
```

```r
sample_means = colMeans(sample_n(Boston_f, 5))
```

```r
for (i in 1:9) {
    sample_means = (sample_means*(i) + colMeans(sample_n(Boston_f, 5)))/(i+1)
}
```

```r
sample_means
```

```r
(sample_means-stats$mean_)/stats$sd_
```
